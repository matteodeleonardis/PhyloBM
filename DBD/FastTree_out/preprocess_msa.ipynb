```julia
using TreeTools, FastaIO
```

```julia
dbd_fasta = readfasta("/home/matteo/Data/DBD/DBD_alignment.uniref90.cov80_clean.a2m")
```

```julia
dbd_sequences = map(x->x[2], dbd_fasta)
```

```julia
dbd_ids = map(x->x[1], dbd_fasta)
```

```julia
unique(dbd_ids) |> length
```

```julia
unique(dbd_sequences) |> length
```

```julia
unique_fasta = []
seen_seqs = []

for (d, s) in dbd_fasta
    if !(s in seen_seqs)
        push!(unique_fasta, (d,s))
        push!(seen_seqs, s)
    end
end
```

```julia
writefasta("DBD_alignment.uniref90.cov80_clean_noduplicates.a2m", unique_fasta)
```
